```python
from glob import glob
import os
from PIL import Image
import numpy as np

base_dir = 'D:/DeepLearn/input/AllInfarct'
label_glob = glob(os.path.join(base_dir, 'LabelDataTotal/*'))
torm_list = []
count = 0
for label_file in label_glob:
    label = Image.open(label_file)
    label_array = np.array(label)
    bol_array = np.array(label) == 2
    if not bol_array.any():
        print(os.path.split(label_file)[-1])
        torm_list.append(os.path.split(label_file)[-1])
        count = count + 1
print(count)
```

## Remove non-infarct files

```python
for fname in torm_list:
    label_f = os.path.join(base_dir, 'LabelDataTotal', fname)
    data_f = os.path.join(base_dir, 'TrainingDataTotal', fname)
    os.remove(label_f)
    os.remove(data_f)
```

## Random sampling

```python
from glob import glob
import os
from PIL import Image
import numpy as np
import random

base_dir = 'D:/DeepLearn/input/AllInfarct'
label_glob = glob(os.path.join(base_dir, 'LabelDataTotal/*'))
data_glob = glob(os.path.join(base_dir, 'TrainingDataTotal/*'))

fnames = [train.split('\\')[-1] for train in data_glob]
if len(fnames)%5 != 0:
    fnum = int(np.ceil(len(fnames)/5) * 5)
else:
    fnum = len(fnames)


training_data = random.sample(fnames, int(fnum*0.8))
training_set = set(training_data)
total_set = set(fnames)
testing_set = total_set - training_set
testing_data = list(testing_set)
```

```python
import shutil
output_dir = base_dir
for train in training_data:
    shutil.copyfile(os.path.join(os.path.join(output_dir, 'TrainingDataTotal', train)), os.path.join(output_dir, 'TrainingData',train))
    shutil.copyfile(os.path.join(os.path.join(output_dir, 'LabelDataTotal', train)), os.path.join(output_dir, 'TrainingLabel',train))
    
for test in testing_data:
    shutil.copyfile(os.path.join(os.path.join(output_dir, 'TrainingDataTotal', test)), os.path.join(output_dir, 'TestingData',test))
    shutil.copyfile(os.path.join(os.path.join(output_dir, 'LabelDataTotal', test)), os.path.join(output_dir, 'TestingLabel',test))
```
